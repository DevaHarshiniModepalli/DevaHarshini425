```python
import statsmodels.formula.api as smf
ml_vol=smf.ols('MPG~VOL',data=cars).fit()
np.round(ml_vol.rsquared,2)
```

**Ordinary least square method!**

R-squared is often interpreted as the "goodness of fit" of a regression model. A higher R-squared value generally indicates a better fit, meaning that the model explains more of the variation in the dependent variable. However, it's important to note that a high R-squared doesn't necessarily mean the model is a good one, and a low R-squared doesn't necessarily mean the model is a bad **one**

```python
import statsmodels.formula.api as smf
ml_vol=smf.ols('MPG~HP',data=cars).fit()
np.round(ml_vol.rsquared,2)
```

```python
import statsmodels.formula.api as smf
ml_vol=smf.ols('MPG~SP',data=cars).fit()
np.round(ml_vol.rsquared,2)
```

```python
import statsmodels.formula.api as smf
ml_vol=smf.ols('MPG~WT',data=cars).fit()
np.round(ml_vol.rsquared,2)
```

```python
import statsmodels.formula.api as smf
ml_vol=smf.ols('MPG~WT+VOL+SP+HP',data=cars).fit()
np.round(ml_vol.rsquared,2)
```
