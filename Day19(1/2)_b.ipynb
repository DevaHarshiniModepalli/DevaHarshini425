```python
from pandas import read_csv
import seaborn as sns
import numpy as np
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2,f_regression,f_classif
dataframe=sns.load_dataset('tips')
dataframe
```

```python
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
dataframe['sex']=lb.fit_transform(dataframe['sex'])
dataframe['smoker']=lb.fit_transform(dataframe['smoker'])
dataframe['day']=lb.fit_transform(dataframe['day'])
dataframe['time']=lb.fit_transform(dataframe['time'])
dataframe
```

```python
X=dataframe.drop('tip',axis=1)
Y=dataframe.tip
```

```python
X
```

```python
Y
```

```python
test=SelectKBest(score_func=f_regression,k=3).fit(X,Y)
```

Well..Let me explain!
Here, **test** is a variable, **SelectKBest** is the method, **score_func** is used to use th tests snd scores are stored. for different purposes, use d**ifferent tes**ts!, k is no.of required best columns that has the best relation to the target coulumn, x is **current** column, y is **target** column
