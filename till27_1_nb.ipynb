```python
def count(t):
  alpha=0
  digit=0
  for i in t:
    if i.isalpha(): # is.alpha() is an inbuilt method of python library
      alpha+=1
    if i.isdigit(): # is.digit() is an inbuilt method of python library
      digit+=1
  return alpha,digit
```
